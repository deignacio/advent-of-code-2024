```python
target = part_1_input
room = target.split('\n')
w = len(room[0])
h = len(room)
start = (-1, -1)
for x in range(w):
    for y in range(h):
        if room[y][x] == '^':
            start = (x, y)

dirs = [(-1, 0), (0, -1), (1, 0), (0, 1)]
dir = 1
```

```python
trail = []
guard = start
while guard[0] >= 0 and guard[1] >= 0 and guard[0] < w and guard[1] < h:
    trail.append(guard)
    next = (guard[0] + dirs[dir][0], guard[1] + dirs[dir][1])
    if next[0] < 0 or next[1] < 0 or next[0] == w or next[1] == h:
        break
    if room[next[1]][next[0]] == '#':
        # turn
        dir += 1
        dir %= len(dirs)
    else:
        guard = next
```

```python
uniq = {}
for t in trail:
    uniq[t] = 1
len(uniq)
```

```python
from collections import defaultdict
def is_loop(room, to_add):
    trail = defaultdict(list)
    guard = start
    dir = 1
    while guard[0] >= 0 and guard[1] >= 0 and guard[0] < w and guard[1] < h:
        if guard in trail and dir in trail[guard]:
            return True
        trail[guard].append(dir)
        next = (guard[0] + dirs[dir][0], guard[1] + dirs[dir][1])
        if next[0] < 0 or next[1] < 0 or next[0] == w or next[1] == h:
            return False
        if room[next[1]][next[0]] == '#' or (next[0] == to_add[0] and next[1] == to_add[1]):
            # turn
            dir += 1
            dir %= len(dirs)
        else:
            guard = next
```

```python
is_loop(room, (3,6))
```

```python
loops = []
for x in range(w):
    for y in range(h):
        if start[0] == x and start[1] == y:
            continue
        if is_loop(room, (x, y)):
            loops.append((x, y))
len(loops)
```
